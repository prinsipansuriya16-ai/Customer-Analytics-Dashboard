```python
from faker import Faker
from datetime import datetime , timedelta 
import random  
import numpy as np
import pandas as pd
import os
```

```python
fake = Faker('en_IN')

customer_num = 50000
detail = []

choices = ['Ahemdabad' , 'Surat' ,'Rajkot','Jamnagar','Gandhinahagr','Bombay','junagath', 'Bangalore', 'Kolkata', 'surendranagar', 'Hyderabad']

for i in range(customer_num):
    detail.append({
        'ID' : 1000+i,
        'name' : fake.name(),
        'city' : random.choice(choices)
    })

df_detail = pd.DataFrame(detail)

indecis = np.random.randint(0,50000,size=1000000)

category = ['electric','cloth','beauty_product','home_kichen']
cat = np.random.choice(category,size=1000000)

quantity = np.random.randint(1,6,size=1000000)
p_name = { 'electric':['fan' ,'fridge','washingmachine' ,'tube light','AC','laptop','mobile','tablet','cooler','TV'] ,'cloth':['shirt' , 'pent' ,'kurti','suit' ,'t shirt','short top'] ,'beauty_product':['foundation' ,'lipstick','blush','powder'] ,
           'home_kichen':['spoon' ,'Dish set' ,'bowl' ,'crockery'] }

product_name = np.array([np.random.choice(p_name[i]) for i in cat])

price_map = {'electric': 50000, 'cloth': 1500, 'beauty_product': 2500, 'home_kichen':500}
base_prices = np.array([price_map[i] for i in cat])


price_variation = np.random.randint(-100,500,size=1000000)
final_price = np.abs(base_prices+price_variation)
```

```python
start_date = datetime(2025 , 11 ,1)

random_date = np.random.randint(0,365,size=1000000)
date_list = [start_date + timedelta(days=int(i)) for i in random_date ]
print(date_list)   
```

```python
df_final_datail = pd.DataFrame({
    "Customer_id": df_detail['ID'].values[indecis],
    "Date" : date_list,
    "Name" : df_detail['name'].values[indecis],
    "City" : df_detail['city'].values[indecis],
    "Product name" : product_name,
    "Category" : cat,
    "Quantity" : quantity,
    "Price per product" : final_price,
    "Total bill" : quantity*final_price
})
```

## save data in csv file

```python
os.makedirs("../data", exist_ok=True)

df_final_datail.to_csv("../data/customer_data.csv", index=False)

sample_df = df_final_datail.sample(
    n=50000,
    random_state=42
)

sample_df.to_csv(
    "../data/sample_customer_data.csv",
    index=False
)

print("Sample dataset saved successfully!")
```
