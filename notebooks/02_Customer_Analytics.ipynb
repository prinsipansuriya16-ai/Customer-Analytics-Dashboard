 # 📊 Customer Analytics Dashboard

## Project Overview

### This project analyzes a self-generated dataset of 1,000,000 customer records using Python, Pandas, and Plotly or matplotlib. The analysis includes sales performance, customer distribution, product category analysis, and business insights.

```python
import numpy as np
import pandas as pd
from datetime import datetime , timedelta 
import matplotlib.pyplot as plt  
import seaborn as sns  
import plotly.graph_objects as go  
```

## 📂 Data Loading


```python
df_final_datail= pd.read_csv("../data/customer_data.csv",parse_dates=["Date"])

```

## 💰 Sales Summary

```python
df_final_datail = df_final_datail.dropna()
df_final_datail
```

## 1.) KPI - Key Performance Indicators

```python
#  Total revenue

#  total bill 

total_bill = df_final_datail["Total bill"].sum()
print("total revenue:",total_bill)

# total order
total_order = len(df_final_datail)
print("total order:",total_order)
    
# AOV - Average order value 
aov = total_bill / total_order
print("Average order value :",aov)

```

## 2.) Monthly Sales Trends

```python

df_final_datail['Year'] = df_final_datail['Date'].dt.year
df_final_datail['Month'] = df_final_datail['Date'].dt.month

monthly_sales = df_final_datail.groupby(by=['Year' , 'Month'])['Total bill'].sum().reset_index()

monthly_sales['Month'] = monthly_sales['Month'].astype(str)
monthly_sales['Year'] = monthly_sales['Year'].astype(str)

month_year=[]
for x,y in zip(monthly_sales['Month'],monthly_sales['Year']):
    month_year.append(f"{x}-{y}")
  
month_year = pd.DataFrame(month_year,columns=['month-year'])
monthly_sales = pd.concat((month_year,monthly_sales),axis=1)


```

```python
# # with the use of matplotlib
# plt.figure(figsize=(12,6))
# sns.pointplot( x=monthly_sales['month-year'] ,y=monthly_sales['Total bill'],marker='*',color='blue')
# plt.title('Month-on-Month Total Sales Trend (November 2025 - November 2026)', fontsize=14, fontweight='bold')
# plt.xlabel('Timeline (Month-Year)', fontsize=12)
# plt.ylabel('Total Revenue (INR)', fontsize=12)
# plt.xticks(rotation=45)
# plt.grid(True, linestyle='--', alpha=0.6)
# plt.tight_layout()
# plt.show()

# with the use of ploty
fig = go.Figure()
fig.add_trace(go.Scatter(
    x=monthly_sales['month-year'],
    y=monthly_sales['Total bill'],
    mode='lines',
    name="Monthly sales",
    line=dict(color='green', width=2)
))
fig.update_layout(
    title='MONTHLY SALES',
    xaxis_title='MONTH-YEAR',
    yaxis_title='$',
    template='plotly_white'
)  

fig.show() 

fig.write_image("../images/monthly_sales.png", width=1200, height=600)
```

## 3.) 🛒 Most Popular Category (Highest Orders)


```python
category_sales = (df_final_datail.groupby("Category")["Quantity"].sum().sort_values(ascending=False))

print(type(category_sales))
print(category_sales)
print("Most popular category : ",category_sales.idxmax())
print("amount of order : ",category_sales.max())
```

```python
fig = go.Figure()

fig.add_trace(go.Bar(
    x = category_sales.index,
    y= category_sales.values,

))
fig.update_layout(
    title='Category wise order',
    xaxis_title='category',
    yaxis_title='order(high-->low)',
    template='plotly_white'
)  
fig.show()
fig.write_image("../images/category_order_sales.png",
                width=1200,
                height=600)
```

## 💰 Top Revenue Generating Category (Highest Sales)

```python
top_revenue = (df_final_datail.groupby('Category')['Total bill'].sum().sort_values(ascending=False))
print(top_revenue)

print("top revenue generating category:",top_revenue.idxmax())
```

```python
fig = go.Figure()

fig.add_trace(go.Bar(
    x = top_revenue.index,
    y= top_revenue.values,

))
fig.update_layout(
    title='Category wise revenue',
    xaxis_title='category',
    yaxis_title='revenue',
    template='plotly_white'
)  
fig.show()
fig.write_image("../images/category_revenue.png",
                width=1200,
                height=600)
```

## 4.) 🌍 Customer Distribution by City

```python
city = pd.DataFrame(df_final_datail['City'].value_counts().reset_index())

print(city.columns)
print(city)
```

```python

fig = go.Figure()
fig.add_trace(go.Scatter(
    x = city['City'],
    y = city['count'],
    name='Customer per city',
    mode='lines',
    line=dict(color='red', width=2, dash='dashdot')
))
fig.update_layout(
    title='Customer per city',
    xaxis_title='city',
    yaxis_title='Number of customer',
    hovermode='x unified',
    template='plotly_white'
)  

fig.show()
fig.write_image("../images/customer_city.png",
                width=1200,
                height=600) 
```
